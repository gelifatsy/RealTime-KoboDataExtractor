```python
import pandas as pd



```

```python
# Specify the path to the CSV file
file_path = '../data/extracted_data.csv'  # Adjust the path if necessary
```

```python

df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
print("\nData types:")
print(df.dtypes)
```

```python
missing_values = df.isnull().sum()
missing_percentages = 100 * df.isnull().sum() / len(df)
missing_table = pd.concat([missing_values, missing_percentages], axis=1, keys=['Missing Values', 'Percentage'])
print("\nMissing values:")
print(missing_table)
```
